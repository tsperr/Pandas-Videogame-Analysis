```python
# Dependencies and Setup
import pandas as pd

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv("Resources/purchase_data.csv")
purchase_data.head()
```

```python
total_players=len(purchase_data["SN"].unique())
purchase_data["Total Players"]=total_players
display(total_players) 
```

```python
unique_items=len(purchase_data["Item Name"].unique())
average_price=round(purchase_data["Price"].mean(axis=0),2)
number_purchases=len(purchase_data["Purchase ID"])
total_revenue=purchase_data["Price"].sum(axis=0)

summary_df=pd.DataFrame({"Number of Unique Items":[unique_items], "Average Price":"$"+ str(average_price), "Number of Purchases":[number_purchases], "Total Revenue": "$"+ str(total_revenue)})

display(summary_df)
```

```python
gender_count=purchase_data["Gender"].value_counts()

male_percentage=round(652/len(purchase_data["Gender"]),2)*100
female_percentage=round(113/len(purchase_data["Gender"]),2)*100
other_percentage=round(15/len(purchase_data["Gender"]),2)*100

summary_df2=pd.DataFrame({"Total Count": gender_count, "Percentage of Players": [str(male_percentage)+ "%", str(female_percentage)+ "%", str(other_percentage)+"%" ]})
summary_df2
```

```python
purchase_count=purchase_data["Gender"].value_counts()

grouped_gender=purchase_data.groupby(["Gender"])
total_purchase_value=grouped_gender["Price"].sum()

average_purchase_price=round(grouped_gender["Price"].mean(),2)

average_person=round((total_purchase_value/purchase_count),2)

summary_df3=pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": average_purchase_price, 
                         "Total Purchase Value": total_purchase_value, "Avg Total Purchase per Person": average_person})
summary_df3
```

```python
bins=[0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 999]
age_bins=["<10","10-14","15-19","20-24","25-29","30-34","35-39", "40+"]

purchase_data["Age Range"]=pd.cut(purchase_data["Age"], bins, labels=age_bins, include_lowest=True)

grouped_age=purchase_data.groupby("Age Range")
total_count=(grouped_age["Item ID"].count())

percentage=round((total_count/total_players)*100,2)
            
summary_df4=pd.DataFrame({"Total Count": total_count, "Percentage of Players": percentage})
summary_df4
```

```python
bins=[0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 999]
age_bins=["<10","10-14","15-19","20-24","25-29","30-34","35-39", "40+"]

purchase_data["Age Ranges"]=pd.cut(purchase_data["Age"], bins, labels=age_bins, include_lowest=True)

grouped_age=purchase_data.groupby("Age Range")
purchase_count=(grouped_age["Item ID"].count())

grouped_age=purchase_data.groupby("Age Range")
average_price=round(grouped_age["Price"].mean(),2)

grouped_age=purchase_data.groupby("Age Range")
purchase_value=round(grouped_age["Price"].sum(),2)

grouped_age=purchase_data.groupby("Age Range")
total_purchase_person=round(grouped_age["Price"].sum()/total_count,2)

summary_df5=pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": average_price, "Total Purchase Value": purchase_value, "Avg Total Purchase per Person": total_purchase_person})
summary_df5
```

```python
grouped_SN=purchase_data.groupby("SN")
purchase_count=(grouped_SN["Item ID"].count())

grouped_SN=purchase_data.groupby("SN")
purchase_price=round(grouped_SN["Price"].mean(),2)

grouped_SN=purchase_data.groupby("SN")
purchase_value=(grouped_SN["Price"].sum())

summary_df6=pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": purchase_price, "Total Purchase Value": purchase_value})
summary_df6
```

```python
popular_df=purchase_data.loc[:, ["Item ID", "Item Name", "Price"]]
popular_df.head()

grouped_ID_name=purchase_data.groupby(["Item ID", "Item Name"])
purchase_count=(grouped_ID_name["Age"].count())

grouped_ID_name=purchase_data.groupby(["Item ID", "Item Name"])
avg_price=round(grouped_ID_name["Price"].mean(),2)

grouped_ID_name=purchase_data.groupby(["Item ID", "Item Name"])
total_value=round(grouped_ID_name["Price"].sum(),2)

summary_df7=pd.DataFrame({"Purchase Count": purchase_count, "Item Price": avg_price, "Total Purchase Value": total_value})
summary_df7=summary_df7.sort_values("Purchase Count", ascending=False)
summary_df7
```
